```python
def caesar_cipher(text, shift):
    result = ""
    
    for char in text:
        # Check if character is a letter
        if char.isalpha():
            # Handle uppercase letters
            if char.isupper():
                # Shift within A–Z range
                new_char = chr((ord(char) - 65 + shift) % 26 + 65)
            else:
                # Shift within a–z range
                new_char = chr((ord(char) - 97 + shift) % 26 + 97)
            result += new_char
        else:
            # Leave numbers, spaces, and punctuation unchanged
            result += char
    
    return result
print(caesar_cipher("Hello, World!", 3))  
print(caesar_cipher("Python", 10))         
print(caesar_cipher("Zebra", 2))           
```

```python

def caesar_cipher(text, shift):
    result = ""
    
    for char in text:
        if char.isalpha():
            base = 65 if char.isupper() else 97
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result


def caesar_decipher(text, shift):
    # Simply reverse the shift
    return caesar_cipher(text, -shift)
encoded = caesar_cipher("Attack at Dawn!", 5)
print(encoded)                # Fyyfhp fy Ifbs!
print(caesar_decipher(encoded, 5))  # Attack at Dawn!
print(caesar_cipher("Hello", 52))   # Hello  (52 wraps around)
print(caesar_cipher("Hello", -3))   # Ebiil

```
