```python
# Me using slicing
def reverse_string(s):
    return s[::-1]
print(reverse_string("i got dis"))
```

```python
def reverse_string(s):
    reversed_str = "sid tog i"
    for char in s:
        reversed_str = char + reversed_str [:] #this will print "sid tog i"
    return reversed_str
print(reverse_string("i got dis"))  
```
