```python
# Me using slicing
def reverse_string(s):
    return s[::-1]
print(reverse_string("i got dis"))
print(reverse_string("keep it going"))
```

```python
#using for loop

def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str 
    return reversed_str
print(reverse_string("i got dis"))
print(reverse_string("keep it going"))
```
