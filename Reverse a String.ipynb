```python
# Me using slicing
def reverse_string(s):
    return s[::-1]
print(reverse_string("i got dis"))
print(reverse_string("keep it going"))
```

```python
#using for loop

def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str 
    return reversed_str
print(reverse_string("i got dis"))
print(reverse_string("keep it going"))
```

```python
#for removing whitespaces
import string 

def reverse_string(s):
    s = ''.join(s.split())  
    return s[::-1]

print(reverse_string("i got dis"))
print(reverse_string("keep it going")) 
```

```python
#for both strings and lists
def reverse_string(s):
    if isinstance(s, str):  
        return s[::-1]
    elif isinstance(s, list):
        return s[::-1]
    else:
        return "Input must be a string or list"

print(reverse_string("i got dis"))
print(reverse_string("keep it going"))   
print(reverse_string([1, 2, 3]))  
```

```python
# timeit comparison
import timeit

def reverse_string_loop(s):
	if isinstance(s, str):
		reversed_str = ""
		for char in s:
			reversed_str = char + reversed_str
		return reversed_str
	elif isinstance(s, list):
		reversed_list = []
		for item in s:
			reversed_list.insert(0, item)
		return reversed_list
	else:
		raise TypeError("Input must be a string or list")

# Measure slicing performance

loop_time = timeit.timeit('reverse_string("i got dis" * 1000000)', globals=globals(), number=100)
print(f"Loop time: {loop_time}")

slicing_time = timeit.timeit('reverse_string("i got dis" * 1000000)', globals=globals(), number=100)
print(f"Slicing time: {slicing_time}")

```
