```python
def find_max(numbers):
   
    if not numbers:
        return None  
    
    max_value = numbers[0]
    
    for num in numbers:
        if num > max_value:
            max_value = num 
    
    return max_value
print(find_max([4, 9, 1, 17, 2]))      
print(find_max([-5, -9, -2, -12]))    
print(find_max([42]))                  
print(find_max([]))                    

```

```python
def find_max(numbers):
    if not numbers:
        return None, None
    
    max_value = numbers[0]
    max_index = 0
    
    for i in range(1, len(numbers)):
        if numbers[i] > max_value:
            max_value = numbers[i]
            max_index = i
    
    return max_value, max_index
print(find_max([4, 9, 1, 17, 2]))
```

```python
def find_max(numbers):
    if not numbers:
        return None
    
    i = 0
    max_value = numbers[0]
    
    while i < len(numbers):
        if numbers[i] > max_value:
            max_value = numbers[i]
        i += 1
    
    return max_value
print(find_max([4, 9, 1, 17, 2]))
```
